# binaryprocessing
- 이진화란 영상을 흑/백으로 분류하여 처리하는 것을 말한다.
- 이때 기준이되는 임계값을 설정하여, 임계값보다 크면 백/작으면 흑으로 표시한다.

```python
import cv2, numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def imshow(tit, image) :
    plt.title(tit)    
    if len(image.shape) == 3 :
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else :
        plt.imshow(image, cmap="gray")
    plt.show()
```

```python
image = cv2.imread('Lena.png',0)
imshow("image", image)
```

## threshold : 임계처리함수
> cv2.threshold(src, thresh[임계값], maxval[임계값을 넘었을 때 적용할 value], type[thresholding type]) 

- THRESH_BINARY : 기준값을 넘으면 최대값 아니면 0
- THRESH_BINARY_INV : 기준값을 넘으면 0 아니면 최대값
- THRESH_TRUNC : 기준값을 넘으면 기준값 아니면 최대값
- THRESH_TOZERO : 기준값을 넘으면 원래값 아니면 0
- THRESH_TOZERO_INV : 기준값을 넘으면 0 아니면 원래값

```python
thr, mask = cv2.threshold(image, 180, 255, cv2.THRESH_BINARY)
imshow("binary", mask)
```

## GausianBlur
> cv2.GaussianBlur(src[이미지], ksize[커널크기], sigmaX[표준편차])


가우시안블러는 가우시안 함수 G를 커널로 사용한다. 가우시안 커널은 중앙위치 (x,y)와 커널위치 (x+u,y+v)의 거리 차이(u,v)가 클수록 가중치가 작아진다.

```python
image = cv2.imread('circle.png',0)    #boxnoise.png
img_blur = cv2.GaussianBlur(image, (21,21), 0)
_, binary = cv2.threshold(img_blur, 90, 255, cv2.THRESH_BINARY)
imshow("", img_blur)
imshow("", binary)
```

## Otsu's Binarization
임계값을 기준으로 영상 픽셀들을 두 클래스로 분류했을 때, 두 클래스간의 intra-class variance를 최소화 혹은 최대화하여 임계값을 찾는 이진화 방법이다. 
<br>
즉, 두개의 봉우리 사이의 값을 선택하여 이진화함으로써 분산을 최소로 만든다.

> 위 코드에서 GaussianBlur를 통해 노이즈를 줄인 후, THRESH_OTSU를 적용해보면 노이즈 없이 물체만 검출된다. 

```python
_, binary = cv2.threshold(img_blur, 0, 255, 
                          cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imshow("", np.hstack([image, img_blur, binary]))

plt.hist(image.ravel(), 256, [0,256], color='r') 
```

## cv2.adaptiveThreshold : 적응 임처리
> cv2.adaptiveThreshold(src, maxValue, adaptiveMethod, thresholdType, blockSize, C)

**Parameters:**	
- src – grayscale image
- maxValue – 임계값
- adaptiveMethod – thresholding value를 결정하는 계산 방법
    - cv2.ADAPTIVE_THRESH_MEAN_C : 주변영역의 평균값으로 결정
    - cv2.ADAPTIVE_THRESH_GAUSSIAN_C : X, Y를 중심으로 block Size * block Size 안에 있는 픽셀 값의 평균에서 C를 뺸 값을 문턱값으로 함
- thresholdType – threshold type
- blockSize – thresholding을 적용할 영역 사이즈
- C - 보정상수(평균이나 가중평균에서 차감할 값) 계산된 값이 양수면 빼주고 음수면 더해준다.

```python
image = cv2.imread('text.jpg',0)   
#img_blur = cv2.GaussianBlur(image, (3,3), 0)
binary = cv2.adaptiveThreshold(image, 255,
          cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
          cv2.THRESH_BINARY, 21, 5)
imshow("", binary)
```

## cv2.medianBlur
kernel window와 pixel의 값들을 정렬한 후에 중간값을 선택하여 적용한다. salt-and-pepper noise(점잡음) 제거에 가장 효과적이다.
> cv2.medianBlur(src, ksize)

**Parameters:**	
- src – 1,3,4 channel image. depth가 CV_8U, CV_16U, or CV_32F 이면 ksize는 3또는5, CV_8U이면 더 큰 ksize가능
- ksize – 1보다 큰 홀수

```python
image = cv2.imread('mart.jpg',0)   
img_blur = cv2.medianBlur(image, 7)
binary = cv2.adaptiveThreshold(img_blur, 255,
          cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
          cv2.THRESH_BINARY, 21, 5)
imshow("", binary)
cv2.imwrite("output.jpg", binary)
```

## 도형의 외곽선 및 꼭지점 찾기

>cv2.findContours(image, mode, method)

contour란 동일한 색 또는 동일한 픽셀값을 가지고 있는 영역의 경계선 정보이다. 물체의 윤곽선 또는 외형을 파악하는데 사용된다.

- **image**: 흑백이미지 또는 이진화된 이미지
- **mode** : 컨투어를 찾는 방법
    - cv2.RETR_EXTERNAL: 컨투어 라인 중 가장 바깥쪽의 라인만 찾음
    - cv2.RETR_LIST: 모든 컨투어 라인을 찾지만, 상하구조(hierachy)관계를 구성하지 않음
    - cv2.RETR_CCOMP: 모든 컨투어 라인을 찾고, 상하구조는 2 단계로 구성함
    - cv2.RETR_TREE: 모든 컨투어 라인을 찾고, 모든 상하구조를 구성함
- **method** : 컨투어를 찾을 때 사용하는 근사화 방법
    - cv2.CHAIN_APPROX_NONE: 모든 컨투어 포인트를 반환
    - cv2.CHAIN_APPROX_SIMPLE: 컨투어 라인을 그릴 수 있는 포인트만 반환
    - cv2.CHAIN_APPROX_TC89_L1: Teh_Chin 연결 근사 알고리즘 L1 버전을 적용하여 컨투어 포인트를 줄임
    - cv2.CHAIN_APPROX_TC89_KCOS: Teh_Chin 연결 근사 알고리즘 KCOS 버전을 적용하여 컨투어 포인트를 줄임

```python
img = cv2.imread('shape4.png',0)
thr, mask = cv2.threshold(img,100,1,cv2.THRESH_BINARY + cv2.THRESH_OTSU)

imshow("",mask)

# findContours는 외곽선(꼭지점)을 찾는 함수. 좌표값으로 나타낸다.
# 즉. countours의 개수를 보고 도형을 예측할 수 있다.
# 3개면 삼각형, 4개면 사각형 ... 등 으로 다각형이라는 것을 1차적으로 예측가능
contours, _ = cv2.findContours(
    mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(contours)
```

## cv2.drawContours

> drawContours(image, contours, contourIdx, color)

- image: 원본 이미지
- contours: 컨투어 라인 정보
- contourIdx: 컨투어 라인 번호

```python
img = cv2.imread('vector.png',0)  # 0으로 해서 흑백영상이 나옴
thr, mask = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
mask = 255 - mask
imshow("",mask)

contours, _ = cv2.findContours(mask, 
                               cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(contours)

image = cv2.imread('vector.png')
cv2.drawContours(image, contours, -1, (0,0,255), 2)
                                   # (색상) 외곽선 굵기

imshow("",image)
```

## cv2.arcLength : 컨투어의 둘레를 구하는 함수. 
<br>
두번째 파라미터의 의미는 폐곡선 여부로, 설정한 값이 
<br>
**True**인 경우 컨투어의 시작점과 끝점을 이어 도형을 구성하고 그 둘레 값을 계산한다. 
<br>
**False**인 경우 시작점과 끝점을 잇지 않고 둘레를 계산한다.

## cv2.contourArea : 컨투어의 면적을 구하는 함수

## approxPolyDP : 컨투어 추정
> cv2.approxPolyDP(cnt, epsilon, True)

- cnt : Numpy Array 형식의 곡선 또는 다각형
- epsilon : 실제 컨투어 라인과 근사치의 최대거리
- True : 폐곡선 여부

```python
font = cv2.FONT_HERSHEY_COMPLEX

img = cv2.imread('shape3.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
thr, mask = cv2.threshold(gray, 0, 255,
        cv2.THRESH_BINARY+cv2.THRESH_OTSU)
mask = 255 - mask
imshow("", mask)
contours, _ = cv2.findContours(mask, 
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for con in contours:
    
    peri = cv2.arcLength(con, True)
    approx = cv2.approxPolyDP(con, 0.02 * peri, True) 
    # 전체 길이에서 2%의 변화가 생기면 approximation 하겠다.
    
    print( f"shape = {len(con)} length={peri}, approx={len(approx)}")
    p = tuple(con[1][0])
    cv2.drawContours(img, [approx], -1, (0,0,255),2)
    cv2.circle(img, p, 5, (255,0,0),-1)
   
    area = cv2.contourArea(con)
    
    if len(approx) == 3:
        cv2.putText(img, "Triangle", p, font, 1, (0))
    elif len(approx) == 4:
        cv2.putText(img, "Rectangle",p , font, 1, (0))
    elif len(approx) == 5:
        cv2.putText(img, "Pentagon",p , font, 1, (0))    
    elif 6 < len(approx) < 15:
        cv2.putText(img, "Ellipse",p , font, 1, (0))
    else:
        cv2.putText(img, "Circle",p , font, 1, (0))
        
imshow("",img)    
```

## 경계면이 붙어있어서 구별이 안되는 이미지
- 팽창시켜서 빈 부분을 뭉치거나
- 약한 연결을 끊어서 **침식** 시킨다. 가장자리를 녹인다.

```python
image = cv2.imread('water_coins.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
binary = 255 - binary

#약한 연결을 침식시키는 연산자 
dilated = cv2.morphologyEx(binary, cv2.MORPH_DILATE, (3,3), iterations=10)
                                            #            침식을 몇번할건지? 

# 팽창시키는 연산자
eroded = cv2.morphologyEx(binary, cv2.MORPH_ERODE, (3,3), iterations=10)             

# 침식 후 팽창
# 원래 크기를 어느정도 유지하면서 주변에 존재하는 노이즈를 없앤다.
opened = cv2.morphologyEx(binary, cv2.MORPH_OPEN,
                          cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5)), iterations=10)             


# 팽창 후 침식
closed = cv2.morphologyEx(binary, cv2.MORPH_CLOSE,
                          cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5)), iterations=10)             

imshow("" , np.hstack([binary,dilated,eroded,opened,closed]))
```

```python
frame = cv2.imread("coin.png")
roi = frame.copy()
gray =  cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
gray_blur = cv2.GaussianBlur(gray, (7, 7), 0)
thresh = cv2.adaptiveThreshold(gray_blur, 255, 
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY_INV, 45, 6)

imshow("", thresh)
```

```python
kernel = np.ones((3, 3), np.uint8)           
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=5)   
imshow("", closing)  
```

```python
cont_img = closing.copy()

contours, _ = cv2.findContours(cont_img, cv2.RETR_EXTERNAL,
cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    area = cv2.contourArea(cnt)
    if area < 2000 or area > 4000:
        continue
    if len(cnt) < 5:
        continue
    ellipse = cv2.fitEllipse(cnt) # 점이 주어지면 타원으로 매핑시키는 함수
    cv2.ellipse(roi, ellipse, (0,255,0), 2)   
    
imshow("out", thresh)                               
imshow("out", cont_img)                               
imshow("out", roi)
```
